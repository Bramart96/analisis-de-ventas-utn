```python
import pandas as pd

url = "https://gist.githubusercontent.com/khanusama20/ee33c2869dd5cf3cebdf020be1ca43f6/raw/cbcbbb2651dd0b631d7bd194bc51b2fbb105d108/sales_sample_2024.csv"

df = pd.read_csv(url)
df.head()
```

importamos datos y dejamos listas las columnas

```python
df.info()
df.columns
```

verificamos que el dataset carga sin errores.

```python
total_ventas = df['sales_amount'].sum()
promedio_ventas = df['sales_amount'].mean()

print("Total ventas:", total_ventas)
print("Promedio ventas:", promedio_ventas)
```

Obtenemos el identificador basico de negocio a traves de metricas

```python
max_venta = df['sales_amount'].max()
min_venta = df['sales_amount'].min()

print("Venta máxima:", max_venta)
print("Venta mínima:", min_venta)
```

detectamos valores maximos y minimos

```python
ventas_por_dia = df.groupby('sales_date')['sales_amount'].sum()
ventas_por_dia
```

analizamos por fecha y obtenemos un comportamiento diario de ventas

```python
import matplotlib.pyplot as plt

ventas_por_dia = df.groupby('sales_date')['sales_amount'].sum()

plt.figure(figsize=(10,5))
plt.plot(ventas_por_dia.index, ventas_por_dia.values)
plt.title("Evolución de ventas por día")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

El gráfico muestra la evolución de las ventas diarias a lo largo del período analizado. Este tipo de análisis permite identificar días con mayor actividad comercial y posibles patrones de consumo.
